1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, maximum):
        if maximum < 0:
            raise ValueError("Maximum cannot be a negative.")
        self.maximum = maximum
        self.value = 0

    def increment(self):
        if self.value < self.maximum:
            self.value += 1
        else:
            print("Error: Counter has reached maximum value.")

    def reset(self):
        self.value = 0

    def get_value(self):
        return self.value
```

```python
def test_counter():
    print("Maximum = 3")
    c = Counter(3)

    print("Initial value:", c.get_value())

    print("Incrementing 3 times:")
    c.increment()
    c.increment()
    c.increment()

    print("Value after 3 increments:", c.get_value())

    print("Attempt to increment beyond maximum:")
    c.increment()  

    print("Value after attempted overflow:", c.get_value()) 

    print("Counter reset:")
    c.reset()

    print("Value after reset:", c.get_value())  

test_counter()
```

```python
# the attempt to increment beyond maximum should be an error
```

2. Copy and paste your solution to question 1 and modify it so that all the data held by the counter is private. Implement functions to check the value of the counter, check the maximum value, and check if the counter is at the maximum.

```python
class Counter:
    def __init__(self, maximum):
        if maximum < 0:
            raise ValueError("Maximum cannot be a negative.")
        self.__maximum = maximum   
        self.__value = 0           

    def increment(self):
        if self.__value < self.__maximum:
            self.__value += 1
        else:
            print("Error: Counter has reached maximum value.")

    def reset(self):
        self.__value = 0

    def get_value(self):
        return self.__value

    def get_maximum(self):
        return self.__maximum

    def is_at_maximum(self):
        return self.__value == self.__maximum
```

```python
def test_counter():
    print("Maximum = 2")
    c = Counter(2)

    print("Initial value:", c.get_value())        
    print("Maximum value:", c.get_maximum())   
    print("Is at maximum?:", c.is_at_maximum())  

    print("\nFirst Increment:")
    c.increment()
    print("Value:", c.get_value())               
    print("Is at maximum?:", c.is_at_maximum()) 

    print("\nSecond Increment:")
    c.increment()
    print("Value:", c.get_value())               
    print("Is at maximum?:", c.is_at_maximum()) 

    print("\nAttempt to increment beyond maximum:")
    c.increment()                               
    print("Value after overflow attempt:", c.get_value())  

    print("\nReset counter:")
    c.reset()
    print("Value after reset:", c.get_value())  
    print("Is at maximum?:", c.is_at_maximum())

test_counter()
```

6. Implement a triangle class analogous to the rectangle and circle in question 5.

```python
import math

class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        if (side1 <= 0 or side2 <= 0 or side3 <= 0):
            raise ValueError("Sides must be positive.")

        if (side1 + side2 <= side3 or
            side1 + side3 <= side2 or
            side2 + side3 <= side1):
            raise ValueError("Invalid triangle sides.")

        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
        self.__x = x
        self.__y = y

    def get_side1(self):
        return self.__side1

    def get_side2(self):
        return self.__side2

    def get_side3(self):
        return self.__side3

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3

    def area(self):
        s = self.perimeter() / 2
        return math.sqrt(
            s *
            (s - self.__side1) *
            (s - self.__side2) *
            (s - self.__side3))
```

```python
def test_triangle():
    print("Triangle: sides 3, 4, 5 and vertex at (0,0)")
    t = Triangle(3, 4, 5, 0, 0)

    print("Sides:", t.get_side1(), t.get_side2(), t.get_side3())
    print("Location:", (t.get_x(), t.get_y()))
    print("Perimeter:", t.perimeter())     
    print("Area:", t.area())              

    try:
        print(t.__side1)
    except AttributeError as e:
        print("Error:", e)


def test_polymorphism():
    shapes = [
        Rectangle(5, 3, 0, 0),
        Circle(4, 1, 2),
        Triangle(3, 4, 5, 2, 2)]

    for shape in shapes:
        print("Shape type:", type(shape).__name__)
        print("Area:", round(shape.area(), 2))
        print("Perimeter:", round(shape.perimeter(), 2))

test_triangle()
print("\nTesting Polymorphism")
test_polymorphism()
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object. 

```python
class Shape:
    def area(self):
        raise NotImplementedError("Subclass must implement area()")

    def perimeter(self):
        raise NotImplementedError("Subclass must implement perimeter()")

    def get_x(self):
        raise NotImplementedError("Subclass must implement get_x()")

    def get_y(self):
        raise NotImplementedError("Subclass must implement get_y()")

    def get_points(self):
        raise NotImplementedError("Subclass must implement get_points()")
```

```python
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        if length <= 0 or width <= 0:
            raise ValueError("Length and width must be positive.")

        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)

    def get_points(self):
        points = []
        steps = 4  

        for i in range(steps):
            points.append((self.__x + i * self.__length / steps, self.__y))
            points.append((self.__x + self.__length,
                           self.__y + i * self.__width / steps))
            points.append((self.__x + self.__length - i * self.__length / steps,
                           self.__y + self.__width))
            points.append((self.__x,
                           self.__y + self.__width - i * self.__width / steps))

        return points[:16]
```

```python
# 4 points per side → total 16
```

```python
import math

class Circle(Shape):
    def __init__(self, radius, x, y):
        if radius <= 0:
            raise ValueError("Radius must be positive.")

        self.__radius = radius
        self.__x = x
        self.__y = y

    def get_radius(self):
        return self.__radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        return math.pi * self.__radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.__radius

    def get_points(self):
        points = []
        for i in range(16):
            angle = 2 * math.pi * i / 16
            px = self.__x + self.__radius * math.cos(angle)
            py = self.__y + self.__radius * math.sin(angle)
            points.append((px, py))
        return points
```

```python
class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        if (side1 <= 0 or side2 <= 0 or side3 <= 0):
            raise ValueError("Sides must be positive.")

        if (side1 + side2 <= side3 or
            side1 + side3 <= side2 or
            side2 + side3 <= side1):
            raise ValueError("Invalid triangle sides.")

        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
        self.__x = x
        self.__y = y

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        s = self.perimeter() / 2
        import math
        return math.sqrt(s *
                         (s - self.__side1) *
                         (s - self.__side2) *
                         (s - self.__side3))

    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3

    def get_points(self):
        x1, y1 = self.__x, self.__y
        x2, y2 = self.__x + self.__side1, self.__y
        x3, y3 = self.__x + self.__side1 / 2, self.__y + self.__side2

        vertices = [(x1, y1), (x2, y2), (x3, y3)]
        points = []

        for i in range(3):
            x_start, y_start = vertices[i]
            x_end, y_end = vertices[(i + 1) % 3]

            for j in range(5):  # 5 points per side ≈ 15 total
                t = j / 5
                px = x_start + t * (x_end - x_start)
                py = y_start + t * (y_end - y_start)
                points.append((px, py))

        return points[:16]
```

```python
def test_points():
    shapes = [
        Rectangle(4, 2, 0, 0),
        Circle(3, 0, 0),
        Triangle(3, 4, 5, 0, 0)]

    for shape in shapes:
        print(type(shape).__name__, "points:")
        pts = shape.get_points()
        print("Number of points:", len(pts))
        print("First 3 points:", pts[:3])

test_points()
```

```python
import math

class Circle(Shape):
    def __init__(self, radius, x, y):
        if radius <= 0:
            raise ValueError("Radius must be positive.")

        self.__radius = radius
        self.__x = x
        self.__y = y

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        return math.pi * self.__radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.__radius

    def contains(self, x, y):
        return ((x - self.__x) ** 2 +
                (y - self.__y) ** 2) <= self.__radius ** 2
```

```python
# (x−x0)^2+(y−y0)^2≤r^2
```

```python
class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        if (side1 <= 0 or side2 <= 0 or side3 <= 0):
            raise ValueError("Sides must be positive.")

        if (side1 + side2 <= side3 or
            side1 + side3 <= side2 or
            side2 + side3 <= side1):
            raise ValueError("Invalid triangle sides.")

        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
        self.__x = x
        self.__y = y

        self.__A = (x, y)
        self.__B = (x + side1, y)
        self.__C = (x + side1 / 2, y + side2)

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3

    def area(self):
        return self.__triangle_area(self.__A, self.__B, self.__C)

    def __triangle_area(self, p1, p2, p3):
        return abs(
            (p1[0]*(p2[1]-p3[1]) +
             p2[0]*(p3[1]-p1[1]) +
             p3[0]*(p1[1]-p2[1])) / 2
        )

    def contains(self, x, y):
        P = (x, y)

        total_area = self.area()
        area1 = self.__triangle_area(P, self.__B, self.__C)
        area2 = self.__triangle_area(self.__A, P, self.__C)
        area3 = self.__triangle_area(self.__A, self.__B, P)

        return abs(total_area - (area1 + area2 + area3)) < 1e-6
```

```python
# Area(ABC) = Area(PAB) + Area(PBC) + Area(PCA)
```

```python
def test_contains():
    shapes = [
        Rectangle(4, 3, 0, 0),
        Circle(3, 0, 0),
        Triangle(4, 3, 5, 0, 0)]

    test_points = [(1, 1), (5, 5), (0, 0)]

    for shape in shapes:
        print("Testing", type(shape).__name__)
        for point in test_points:
            print("Point", point, "inside?",
                  shape.contains(point[0], point[1]))

test_contains()
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
class Shape:
    def area(self):
        raise NotImplementedError("Subclass must implement area()")

    def perimeter(self):
        raise NotImplementedError("Subclass must implement perimeter()")

    def get_x(self):
        raise NotImplementedError("Subclass must implement get_x()")

    def get_y(self):
        raise NotImplementedError("Subclass must implement get_y()")

    def get_points(self):
        raise NotImplementedError("Subclass must implement get_points()")

    def contains(self, x, y):
        raise NotImplementedError("Subclass must implement contains(x, y)")

    def overlaps(self, other):
        raise NotImplementedError("Subclass must implement overlaps(other)")
```

```python
class Rectangle(Shape):

    def overlaps(self, other):
        x1_min = self.__x
        x1_max = self.__x + self.__length
        y1_min = self.__y
        y1_max = self.__y + self.__width

        other_points = other.get_points()
        x2_min = min(p[0] for p in other_points)
        x2_max = max(p[0] for p in other_points)
        y2_min = min(p[1] for p in other_points)
        y2_max = max(p[1] for p in other_points)

        return not (x1_max < x2_min or x1_min > x2_max or
                    y1_max < y2_min or y1_min > y2_max)
```

```python
import math

class Circle(Shape):

    def overlaps(self, other):
        other_points = other.get_points()
        x2_min = min(p[0] for p in other_points)
        x2_max = max(p[0] for p in other_points)
        y2_min = min(p[1] for p in other_points)
        y2_max = max(p[1] for p in other_points)

        closest_x = max(x2_min, min(self.__x, x2_max))
        closest_y = max(y2_min, min(self.__y, y2_max))

        distance = math.sqrt((self.__x - closest_x) ** 2 + (self.__y - closest_y) ** 2)
        return distance <= self.__radius
```

```python
class Triangle(Shape):

    def overlaps(self, other):
        xs = [self.__A[0], self.__B[0], self.__C[0]]
        ys = [self.__A[1], self.__B[1], self.__C[1]]
        x1_min, x1_max = min(xs), max(xs)
        y1_min, y1_max = min(ys), max(ys)

        other_points = other.get_points()
        x2_min = min(p[0] for p in other_points)
        x2_max = max(p[0] for p in other_points)
        y2_min = min(p[1] for p in other_points)
        y2_max = max(p[1] for p in other_points)

        return not (x1_max < x2_min or x1_min > x2_max or
                    y1_max < y2_min or y1_min > y2_max)
```

```python
def test_overlaps():
    r1 = Rectangle(4, 3, 0, 0)
    r2 = Rectangle(2, 2, 1, 1)
    c1 = Circle(2, 6, 6)
    t1 = Triangle(3, 4, 5, 3, 0)

    shapes = [r1, r2, c1, t1]

    for i, s1 in enumerate(shapes):
        for j, s2 in enumerate(shapes):
            if i < j:
                print(f"{type(s1).__name__} overlaps {type(s2).__name__}?",
                      s1.overlaps(s2))

test_overlaps()
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

```python
import math

class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]

    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]

    def display(self):
        print("\n".join(["".join(row) for row in self.data]))

class Shape:
    def area(self):
        raise NotImplementedError
    def perimeter(self):
        raise NotImplementedError
    def get_x(self):
        raise NotImplementedError
    def get_y(self):
        raise NotImplementedError
    def get_points(self):
        raise NotImplementedError
    def contains(self, x, y):
        raise NotImplementedError
    def overlaps(self, other):
        raise NotImplementedError
    def paint(self, canvas, char='*'):
        for px, py in self.get_points():
            canvas.set_pixel(int(py), int(px), char)

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def area(self): return self.__length * self.__width
    def perimeter(self): return 2*(self.__length + self.__width)
    def contains(self, x, y):
        return self.__x <= x <= self.__x+self.__length and self.__y <= y <= self.__y+self.__width
    def get_points(self):
        points = []
        # simple 16 points along edges
        for i in range(4):
            points.append((self.__x + i * self.__length/3, self.__y))
            points.append((self.__x + self.__length, self.__y + i * self.__width/3))
            points.append((self.__x + self.__length - i * self.__length/3, self.__y + self.__width))
            points.append((self.__x, self.__y + self.__width - i * self.__width/3))
        return points[:16]
    def overlaps(self, other):
        other_points = other.get_points()
        x1_min, x1_max = self.__x, self.__x+self.__length
        y1_min, y1_max = self.__y, self.__y+self.__width
        x2_min = min(p[0] for p in other_points)
        x2_max = max(p[0] for p in other_points)
        y2_min = min(p[1] for p in other_points)
        y2_max = max(p[1] for p in other_points)
        return not (x1_max < x2_min or x1_min > x2_max or y1_max < y2_min or y1_min > y2_max)

class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def area(self): return math.pi * self.__radius**2
    def perimeter(self): return 2*math.pi*self.__radius
    def contains(self, x, y):
        return (x-self.__x)**2 + (y-self.__y)**2 <= self.__radius**2
    def get_points(self):
        points = []
        for i in range(16):
            angle = 2*math.pi*i/16
            px = self.__x + self.__radius*math.cos(angle)
            py = self.__y + self.__radius*math.sin(angle)
            points.append((px, py))
        return points
    def overlaps(self, other):
        other_points = other.get_points()
        x2_min = min(p[0] for p in other_points)
        x2_max = max(p[0] for p in other_points)
        y2_min = min(p[1] for p in other_points)
        y2_max = max(p[1] for p in other_points)
        closest_x = max(x2_min, min(self.__x, x2_max))
        closest_y = max(y2_min, min(self.__y, y2_max))
        distance = math.sqrt((self.__x - closest_x)**2 + (self.__y - closest_y)**2)
        return distance <= self.__radius

class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
        self.__x = x
        self.__y = y
        self.__A = (x, y)
        self.__B = (x+side1, y)
        self.__C = (x+side1/2, y+side2)

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def perimeter(self): return self.__side1+self.__side2+self.__side3
    def area(self):
        x1,y1 = self.__A
        x2,y2 = self.__B
        x3,y3 = self.__C
        return abs((x1*(y2-y3)+x2*(y3-y1)+x3*(y1-y2))/2)
    def contains(self, x, y):
        px, py = x, y
        x1,y1 = self.__A
        x2,y2 = self.__B
        x3,y3 = self.__C
        area_total = self.area()
        area1 = abs((px*(y2-y3)+x2*(y3-py)+x3*(py-y2))/2)
        area2 = abs((x1*(py-y3)+px*(y3-y1)+x3*(y1-py))/2)
        area3 = abs((x1*(y2-py)+x2*(py-y1)+px*(y1-y2))/2)
        return abs(area_total - (area1+area2+area3))<1e-6
    def get_points(self):
        points = []
        vertices = [self.__A, self.__B, self.__C]
        for i in range(3):
            x_start,y_start = vertices[i]
            x_end,y_end = vertices[(i+1)%3]
            for j in range(5):
                t = j/5
                px = x_start + t*(x_end-x_start)
                py = y_start + t*(y_end-y_start)
                points.append((px, py))
        return points[:16]
    def overlaps(self, other):
        xs = [self.__A[0], self.__B[0], self.__C[0]]
        ys = [self.__A[1], self.__B[1], self.__C[1]]
        x1_min, x1_max = min(xs), max(xs)
        y1_min, y1_max = min(ys), max(ys)
        other_points = other.get_points()
        x2_min = min(p[0] for p in other_points)
        x2_max = max(p[0] for p in other_points)
        y2_min = min(p[1] for p in other_points)
        y2_max = max(p[1] for p in other_points)
        return not (x1_max<x2_min or x1_min>x2_max or y1_max<y2_min or y1_min>y2_max)

class CompoundShape(Shape):
    def __init__(self):
        self.shapes = []
    def add_shape(self, shape):
        self.shapes.append(shape)
    def area(self):
        return sum(s.area() for s in self.shapes)
    def perimeter(self):
        return sum(s.perimeter() for s in self.shapes)
    def get_points(self):
        points=[]
        for s in self.shapes:
            points.extend(s.get_points())
        return points[:16]
    def contains(self, x, y):
        return any(s.contains(x,y) for s in self.shapes)
    def overlaps(self, other):
        return any(s.overlaps(other) for s in self.shapes)
    def paint(self, canvas, char='*'):
        for s in self.shapes:
            s.paint(canvas,char)
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def remove_shape(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)

    def paint(self):
        self.canvas.clear_canvas()
        for shape in self.shapes:
            shape.paint(self.canvas)
        self.canvas.display()

    def clear(self):
        self.shapes = []
        self.canvas.clear_canvas()
```

```python
if __name__ == "__main__":
    drawing = RasterDrawing(40, 20)

    rect = Rectangle(10, 4, 2, 2)
    circ = Circle(3, 25, 5)
    tri = Triangle(6, 4, 5, 15, 10)

    drawing.add_shape(rect)
    drawing.add_shape(circ)
    drawing.add_shape(tri)

    print("Initial drawing:")
    drawing.paint()  

    new_rect = Rectangle(5, 3, 5, 15)
    drawing.add_shape(new_rect)

    print("\nModified drawing (added new rectangle):")
    drawing.paint()  

    drawing.remove_shape(circ)

    print("\nModified drawing (removed circle):")
    drawing.paint() 
```

   * Overload `__repr__` functions of all objects to return strings of the python code that would construct the object.
   
   * In the save method of raster drawing class, store the representations into the file.
   * Write a loader function that reads the file and uses `eval` to instantiate the object.

For example:

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```

```python
class Rectangle(Shape):
    def __repr__(self):
        return f"Rectangle({self.__length}, {self.__width}, {self.__x}, {self.__y})"

class Circle(Shape):
    def __repr__(self):
        return f"Circle({self.__radius}, {self.__x}, {self.__y})"

class Triangle(Shape):
    def __repr__(self):
        return f"Triangle({self.__side1}, {self.__side2}, {self.__side3}, {self.__x}, {self.__y})"

class CompoundShape(Shape):
    def __repr__(self):
        shapes_repr = ",".join(repr(s) for s in self.shapes)
        return f"CompoundShape([{shapes_repr}])"
```

```python
class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def remove_shape(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)

    def paint(self):
        self.canvas.clear_canvas()
        for shape in self.shapes:
            shape.paint(self.canvas)
        self.canvas.display()

    def clear(self):
        self.shapes = []
        self.canvas.clear_canvas()

    def __repr__(self):
        shapes_code = "[" + ",".join(repr(s) for s in self.shapes) + "]"
        return f"RasterDrawing({self.canvas.width}, {self.canvas.height}, shapes={shapes_code})"

    def save(self, filename):
        with open(filename, "w") as f:
            f.write(self.__repr__())

def raster_loader(filename):
    with open(filename, "r") as f:
        obj = eval(f.read())
    return obj
```

```python
if __name__ == "__main__":
    drawing = RasterDrawing(40, 20)
    r = Rectangle(10, 4, 2, 2)
    c = Circle(3, 25, 5)
    t = Triangle(6, 4, 5, 15, 10)

    drawing.add_shape(r)
    drawing.add_shape(c)
    drawing.add_shape(t)

    print("Original Drawing:")
    drawing.paint()

    drawing.save("my_drawing.rdr")
    loaded_drawing = raster_loader("my_drawing.rdr")

    print("\nLoaded Drawing:")
    loaded_drawing.paint()
```
